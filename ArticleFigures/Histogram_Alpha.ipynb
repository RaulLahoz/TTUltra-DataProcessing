### Notebook $\alpha$ 

Notebook para realizar un histograma de los diferentes valores obtenidos para $\alpha$. Tomaremos los diferemos valores de $\alpha$ obtenidos para cada una de nuestras mediciones y haremos un histograma demostrando la distribución de estos valores, los cuales esperamos que estén en el rango de 0 < |$\alpha$| < 1.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Diccionario con valores de alpha
alpha_ON = {
    "C60_15_CW": 0.7574,
    "C60_18_CW_1": 0.1570,
    "C60_18_CW_2": 0.2425,
    "C60_19_CW": 0.3698,
    "C60_23_CW": 0.9481,
    "C60_29_CW": 0.2986,
    "C60_13_CW": 0.5910,
    "C60_12_CW": 0.1251,
    "C60_9_CW": 0.0396,
    "C60_8_CW": 0.5216,
}
```

```python
# Diccionario con valores de los errores de alpha
alpha_ON_err = {
    "C60_15_CW": 0.0165,
    "C60_18_CW_1": 0.0277,
    "C60_18_CW_2": 0.0235,
    "C60_19_CW": 0.0112,
    "C60_23_CW": 0.0428,
    "C60_29_CW": 0.0248,
    "C60_13_CW": 0.0079,
    "C60_12_CW": 0.0383,
    "C60_9_CW": 0.0131,
    "C60_8_CW": 0.0123,
}
```

```python
# Número de bins
n_bins = 10  # Puedes cambiar este valor

# Tamaño de fuente
font_size = 16

# Extraer valores de alpha
alpha_values = list(alpha_ON.values())

# Crear el histograma
plt.figure(figsize=(6, 4))
counts, bins, patches = plt.hist(alpha_values, bins=n_bins, color='black', edgecolor='black', alpha=0.7)

# Calcular los centros de los bins y formatearlos con 2 decimales
bin_centers = (bins[:-1] + bins[1:]) / 2
bin_labels = [f"{center:.2f}" for center in bin_centers]

# Etiquetas y título con tamaño de fuente
plt.xlabel(r'$\alpha$ values', fontsize=font_size)
plt.ylabel("Frequency", fontsize=font_size)

# Ajustar etiquetas del eje X con los centros de los bins y rotarlos 45 grados
plt.xticks(bin_centers, bin_labels, fontsize=font_size, rotation=45)

# Ajustar el eje Y para que solo muestre números enteros
plt.yticks(np.arange(0, max(counts) + 1, 1), fontsize=font_size)

# Mostrar la gráfica
plt.show()
```
