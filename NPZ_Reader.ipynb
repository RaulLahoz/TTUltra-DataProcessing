### Post-analysis using .npz files

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load("hist.npz")
```

```python
xcorr = data["arr_0"]
ycorr = data["arr_1"]
ycorr_norm = data["arr_2"]

plt.plot(xcorr, ycorr)
plt.show()
plt.plot(xcorr, ycorr_norm)
plt.show()
```
